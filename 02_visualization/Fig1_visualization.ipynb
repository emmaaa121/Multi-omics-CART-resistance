```python
import matplotlib.pyplot as plt
import numpy as np
import os
import scanpy as sc
from matplotlib.lines import Line2D

def plot_gene_expression(embedding, categories, color_map, xlabel, ylabel, plot_title, file_name):
    plt.figure(figsize=(12, 10))
    plt.rcParams.update({'font.size': 30})
    ax = plt.gca()
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    unique_categories = np.unique(categories)
    for cat in unique_categories:
        indices = np.where(categories == cat)[0]
        cluster_points = embedding[indices, :]
        plt.scatter(cluster_points[:, 0], cluster_points[:, 1], 
                    label=cat, marker='o', s=8, alpha=0.6, color=color_map[cat])  

    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(plot_title, fontsize=35, fontweight='bold')

    custom_legend = [Line2D([0], [0], marker='o', color='w', label='CD4', markersize=10, markerfacecolor='#E9B872', alpha=0.6),
                     Line2D([0], [0], marker='o', color='w', label='CD8', markersize=10, markerfacecolor='#6494AA', alpha=0.6)]
    plt.legend(handles=custom_legend, loc='lower right', markerscale=2, frameon=False)

    plt.tight_layout()

    result_dir = '/home/emma/result/CART/'
    if not os.path.exists(result_dir):
        os.makedirs(result_dir)
    
    save_path = os.path.join(result_dir, file_name)
    plt.savefig(save_path, dpi=300)
    plt.show()

adata = sc.read_h5ad('/home/emma/data/CART/Harvard_Stanford_infusion_D7sorted_CD3E_CD4_CD8A_highly_variable_harmony_genovis.h5ad') 
print(adata.shape)

# Extract the expression values for CD4 and CD8A
CD4_expression = adata[:, 'CD4'].X.toarray().flatten()
CD8_expression = adata[:, 'CD8A'].X.toarray().flatten()
categories = np.where(CD8_expression > CD4_expression, 'CD8', np.where(CD4_expression > CD8_expression, 'CD4', 'NA'))

color_map = {
    'CD4': '#E9B872',
    'CD8': '#6494AA',
    'NA': 'gray'
}

plot_gene_expression(
    adata.obsm['X_genovis'], 
    categories, 
    color_map, 
    'Genovis1', 
    'Genovis2', 
    '', 
    'CD4_CD8_Expression_genovis.png'
)
```
